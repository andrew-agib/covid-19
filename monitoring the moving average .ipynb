```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
df = pd.read_excel(r'C:\Users\49173\Downloads\covid19.xlsx')
df.head()
#df=df.set_index('DateRep')
```

```python
value=0
j=1
cases= pd.DataFrame()
cases=cases.fillna(0)
count=[]
for i in range(0,len(df)-1):
    if df.loc[i,"Countries and territories"]==df.loc[i+1,"Countries and territories"]:
        count.append(df.loc[i,"Cases"])
        value=value+1
    else:
        cases.loc[0,j]=df.loc[i,"Countries and territories"]
        for q in range(1,value):
            cases.loc[q,j]=count[q]
        j=j+1
        value=0
        count=[]

```
